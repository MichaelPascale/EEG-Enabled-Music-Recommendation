# Filter CSV
Filter the features CSV file to remove all features other than MFCC and all tracks other than those of interest.

```python
import pandas

mfcc = pandas.read_csv("mfcc.csv")
mfcc
```

```python
tracks = pandas.read_csv("raw_tracks_edited.csv")
tracks
```

```python
type(mfcc)
type(tracks)
```

```python
# We only want MFCC for tracks listed in raw_tracks_edited.csv
track_ids = tracks['track_id'].to_list()
track_ids.sort()
track_ids[:5] + track_ids[-5:]
```

```python
new_mfcc = mfcc[mfcc.track_id.isin(track_ids)]
new_mfcc
```

```python
new_mfcc = new_mfcc.sort_values('track_id')
tracks = tracks.sort_values('track_id')
new_mfcc.head()
#tracks.head()
```

```python
tracks.shape
```

```python
new_mfcc.shape
```

```python
nmfcc = new_mfcc.copy()
```

```python
nmfcc = nmfcc.merge(tracks[['track_id', 'genre']], on='track_id')
nmfcc
```

```python
nnmfcc = nmfcc.copy()
cols = list(nnmfcc.columns.values)
cols.pop(cols.index('genre'))
cols.pop(cols.index('track_id'))
nnmfcc = nnmfcc[['track_id', 'genre'] + cols]
```

```python
nnmfcc
```

```python
nnmfcc.to_csv('output.csv')
```
