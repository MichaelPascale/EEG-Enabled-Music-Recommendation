```python
import numpy as np
from tensorflow import keras
```

```python
import pandas
input_data = pandas.read_csv("preprocessed_mfccs.csv")
input_data.head()
```

```python
mapper = {"classical": 0, "jazz": 1, "metal": 2, "rock": 3, "pop":4 }
labels = input_data['genre'].replace(mapper)
labels.shape
```

```python
labels = keras.utils.to_categorical(labels)
labels.shape
```

```python
data = input_data.drop(columns=['genre'])
data.shape
```

```python
v = data.values
# For each column of v, sum all over the rows to get the column total.
cmxn = [(max([abs(r[c]) for r in v]), min([abs(r[c]) for r in v])) for c in range(len(v[0]))]
normalized = [[(r[c]-cmxn[c][1])/(cmxn[c][0]-cmxn[c][1]) for r in v] for c in range(len(v[0]))]
normalized = pandas.DataFrame(normalized).T
normalized
```

```python
model = keras.Sequential()
# Feature Layer + 1 Hidden Layer
model.add(keras.layers.BatchNormalization(input_shape=(data.shape[1],)))
model.add(keras.layers.Dense(20, activation='sigmoid'))
model.add(keras.layers.BatchNormalization())
# Classification Layer
model.add(keras.layers.Dense(labels.shape[1], activation='sigmoid'))
# Optimizer
sgd = keras.optimizers.SGD(lr=0.3, momentum=0.2, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
history = model.fit(data, labels, epochs=500, shuffle=True, validation_split=0.33)
```

```python
model.save('tfmodel41')
```
